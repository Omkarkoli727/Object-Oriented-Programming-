```python
class ATM:
    def __init__(self,bank_acc,balance,atm_pin,card_blocked=False,wrong_pin_attempts=0):
        self.bank_acc=bank_acc
        self.balance=balance
        self.atm_pin=atm_pin
        self.card_blocked=card_blocked
        self.wrong_pin_attempts=wrong_pin_attempts
    def check_balance(self,atm_pin):
        if self.atm_pin==atm_pin and self.card_blocked==False:
            print(f"Account Number : {self.bank_acc}")
            print(f"Balance = {self.balance}")
        elif self.card_blocked==True:
            print(f"your card is blocked!!...")
        else: 
            self.wrong_pin_attempts+=1
            print(f"wrong pin...")
            if self.wrong_pin_attempts>=3:
                self.card_blocked=True
                print(f"your card is blocked due to 3 wrong pin attempts!!...")
    def deposit(self,amount,atm_pin):
        if self.atm_pin==atm_pin and self.card_blocked==False:
            self.balance=self.balance+amount
            print(f"amount added on your account!!...")
            print(f"total balance = {self.balance}")
        elif self.card_blocked==True:
            print(f"your card is blocked!!...")
        else: 
            self.wrong_pin_attempts+=1
            print(f"wrong pin...")
            if self.wrong_pin_attempts>=3:
                self.card_blocked=True
                print(f"your card is blocked due to 3 wrong pin attempts!!...")
    def withdraw(self,amount,atm_pin):
        if amount>self.balance:
            print(f"insufficient balance!!...")
        elif self.atm_pin==atm_pin and self.card_blocked==False :
            self.balance=self.balance-amount
            print(f"amount withdraw from your account!!...")
            print(f"total balance = {self.balance}")
        elif self.card_blocked==True:
            print(f"your card is blocked!!...")
        else: 
            self.wrong_pin_attempts+=1
            print(f"wrong pin...")
            if self.wrong_pin_attempts>=3:
                self.card_blocked=True
                print(f"your card is blocked due to 3 wrong pin attempts!!...")
    def change_pin(self,old_pin,new_pin):
        if self.atm_pin==old_pin and self.card_blocked==False:
            self.atm_pin=new_pin
            print(f"your pin is successfully changed!!...")
        elif self.card_blocked==True:
            print(f"your card is blocked!!...")
        else: 
            self.wrong_pin_attempts+=1
            print(f"wrong old pin...")
            if self.wrong_pin_attempts>=3:
                self.card_blocked=True
                print(f"your card is blocked due to 3 wrong pin attempts!!...")
    def transfer(self,amount,atm_pin,receiver_acc):
        if self.atm_pin==atm_pin and self.card_blocked==False:
            self.balance=self.balance-amount
            print(f"amount transfer from your account!!...")
            print(f"total balance = {self.balance}")
            print(f"amount transfer to account number : {receiver_acc}")
        elif self.card_blocked==True:
            print(f"your card is blocked!!...")
        else: 
            self.wrong_pin_attempts+=1
            print(f"wrong pin...")
            if self.wrong_pin_attempts>=3:
                self.card_blocked=True
                print(f"your card is blocked due to 3 wrong pin attempts!!...")
    def mini_statement(self,atm_pin):
        if self.atm_pin==atm_pin and self.card_blocked==False:
            print(f"Account Number : {self.bank_acc}")
            print(f"Balance = {self.balance}")
            print(f"last 5 transactions : ...")
        elif self.card_blocked==True:
            print(f"your card is blocked!!...")
        else: 
            self.wrong_pin_attempts+=1
            print(f"wrong pin...")
            if self.wrong_pin_attempts>=3:
                self.card_blocked=True
                print(f"your card is blocked due to 3 wrong pin attempts!!...")
    def block_card(self,atm_pin):
        if self.atm_pin==atm_pin:
                self.card_blocked=True
                print(f"your card is blocked!!...")
        else: 
            self.wrong_pin_attempts+=1
            print(f"wrong pin...")
            if self.wrong_pin_attempts>=3:
                self.card_blocked=True
                print(f"your card is blocked due to 3 wrong pin attempts!!...")
    def unblock_card(self,atm_pin):
        if self.atm_pin==atm_pin:
                self.card_blocked=False
                self.wrong_pin_attempts=0
                print(f"your card is unblocked!!...")
        else: 
            self.wrong_pin_attempts+=1
            print(f"wrong pin...")
            if self.wrong_pin_attempts>=3:
                self.card_blocked=True
                print(f"your card is blocked due to 3 wrong pin attempts!!...")
    
    
```
